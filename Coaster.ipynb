```python
df = pd.read_csv('coaster_db.csv')
```

```python
#Subsetting of data columns 
# we can use drop method as well to do it 
# df.drop(['column_name'],axis =1 )

df = df[['coaster_name', 
    #'Length', 'Speed', 
    'Location', 'Status', 
    #'Opening date',
       #'Type', 
    'Manufacturer', #'Height restriction', 'Model', 'Height',
      # 'Inversions', 'Lift/launch system', 'Cost', 'Trains', 'Park section',
     #  'Duration', 'Capacity', 'G-force', 'Designer', 'Max vertical angle',
     #  'Drop', 'Soft opening date', 'Fast Lane available', 'Replaced',
     #  'Track layout', 'Fastrack available', 'Soft opening date.1',
      # 'Closing date', 'Opened', 'Replaced by', 'Website',
      # 'Flash Pass Available', 'Must transfer from wheelchair', 'Theme',
      # 'Single rider line available', 'Restraint Style',
      # 'Flash Pass available', 'Acceleration', 'Restraints', 'Name',
       'year_introduced',
    'latitude', 'longitude', 'Type_Main',
     'opening_date_clean', 
    #'speed1', 'speed2', 'speed1_value', 'speed1_unit',
       'speed_mph', #'height_value', 'height_unit',
    'height_ft',
       'Inversions_clean', 'Gforce_clean']].copy()
```

```python
df.shape
```

```python
df['opening_date_clean'] =  pd.to_datetime(df['opening_date_clean'])
```

```python
df.dtypes
```

```python
# Rename of Columns
df = df.rename(columns={'coaster_name':'Coaster_Name','year_introduced':'Year_Introduced','latitude':'Latitude','longitude':'Longitude','opening_date_clean':
                   'Opening_Date','speed_mph':'Speed_mph','height_ft':'Height_ft'})
```

```python
# To see the count of how many null values present in each column
df.isna().sum()
```

```python
# To see how manyg rows are duplicated 
df.loc[df.duplicated()]
```

```python
df.loc[df.duplicated(subset=['Coaster_Name'])]
```

```python
#Checking for a duplicate with query 
df.query('Coaster_Name == "Crystal Beach Cyclone"')
```

```python
df.loc[~df.duplicated(subset=['Coaster_Name', 'Location' ,'Opening_Date'])]
```

```python
# index is used to reset the rows which can be used if we delete anykind of rows in df
# \ it is used to break the lines of codes 
df = df.loc[~df.duplicated(subset=['Coaster_Name', 'Location' ,'Opening_Date'])] \
    .reset_index(drop=True).copy()
```

```python
df.shape
```

```python
df.isnull().sum()
```

```python
# Using mode to fill the null values
df[df['Status'].isnull()]
```

```python
df.head(20)
```

```python
mode=''.join(df['Status'].mode())
df['Status'].fillna(mode,inplace=True)
```

```python
# to fill null values for Manufacture column by using ffill method because it is a non-numeric column
df['Manufacturer'].fillna(method='ffill',inplace=True)
```

```python
df.isna().sum()
```

```python
#To fill null values using mean for more than one column
df[['Speed_mph', 'Height_ft']] = df[['Speed_mph', 'Height_ft']].fillna(df[['Speed_mph', 'Height_ft']].mean())
#one more mwthod using other than pandas library by SimpleImputer from Scikit-learn
#from sklearn.impute import SimpleImputer

#imputer = SimpleImputer(strategy='mean')  # Fill NaNs using column mean
#df[['column1', 'column2']] = imputer.fit_transform(df[['column1', 'column2']])
```

```python
df.isna().sum()
```

```python
#using bfill for Opening values to fill NaT(Not a Time) values
df['Opening_Date'].fillna(method='bfill',inplace=True)
```

```python
df.dtypes
```

```python
# Using values 0 to fill null values
df['Gforce_clean'].fillna(0,inplace=True)
```

```python
#using median to fill na values for ccolumn-Latitute
df['Latitude'].fillna(df['Latitude'].median(),inplace=True)
```

```python
#Using mean to fill na values for column-Longitude
df['Longitude'].fillna(df['Longitude'].mean(),inplace=True)
```

```python
df.isna().sum()
```

```python
df['Year_Introduced'].value_counts()
```

```python
# Value counts shows how many times each value is repeated in the columns
a = df['Year_Introduced'].value_counts().head(10).plot(kind='bar' ,title="Top 10 Years Coasters Introducted")
a.set_xlabel('Year_Introducted')
a.set_ylabel('Count')
```

```python
#histogram groups data in to bins and show the count frequency values in each bin.
b = df['Speed_mph'].plot(kind='hist' ,bins=20,title= 'Coaster speed(mph)')
b.set_xlabel('Speed(mph)')
```

```python
# kde plot estimates the underlying distibution of data using a smooth curve 
b = df['Speed_mph'].plot(kind='kde',title= 'Coaster speed(mph)')
b.set_xlabel('Speed(mph)')
```

```python
# by adding the plt.show() it removes the subplot(unwanted data  
df.plot(kind='scatter',x="Speed_mph",y="Height_ft",title='Coaster Speed Vs Height')
plt.show()
```

```python
#scatter plot using seaborn 
# hue parameter is used to differeniate data by using colours
ax= sns.scatterplot(x="Speed_mph",y="Height_ft",
                hue='Year_Introduced', data=df)
ax.set_title('Coaster Speed Vs Height')
plt.show()
```

```python
#pairplot used to comapre multiple features in the df
sns.pairplot(df,vars=['Year_Introduced','Speed_mph','Height_ft','Inversions_clean','Gforce_clean'],hue='Type_Main')
plt.show()
```

```python
df_corr = df[['Year_Introduced','Speed_mph','Height_ft','Inversions_clean','Gforce_clean']].dropna().corr()
df_corr
```

```python
# to find correlation with graph using heatmap in seaborn 
sns.heatmap(df_corr,annot=True)
```

```python
# to find top 10 loactions with speed of coaster's.
ax = df.query('Location!="Other"').groupby('Location')['Speed_mph'].agg(['mean','count']) \
    .query('count>=10').sort_values('mean')['mean'].plot(kind='barh',title='Average Coaster Speed by Location')
ax.set_xlabel('Average Coaster Speed')
plt.show()
```
